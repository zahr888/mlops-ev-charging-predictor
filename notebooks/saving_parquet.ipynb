```python
import boto3
import pandas as pd
df=pd.read_parquet("s3://ev-data/parquets/features.parquet", storage_options = {
    "client_kwargs": {"endpoint_url": "http://localhost:4566"},
    "key": "test",
    "secret": "test"
})

split_date = df.index[-24*30]
test_df = df[df.index >= split_date]

s3 = boto3.client("s3", endpoint_url="http://localhost:4566", aws_access_key_id="test", aws_secret_access_key="test")
s3.put_object(Bucket="ev-data", Key="parquets/test_features.parquet", Body=test_df.to_parquet())
print("Test data uploaded to S3 successfully.")
```
